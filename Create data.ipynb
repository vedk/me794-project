```python
import pandas as pd
```

```python
mean_wear = []
mean_friction = []
for i in range(1, 97):
    df = pd.read_csv(f"Data/Piezo_{i}.dwf", sep=";", names=["timestamp", "wear", "friction", "const", "ignore"])
    mean_wear.append(df["wear"].loc[10:60].mean())
    mean_friction.append(df["friction"].loc[10:60].mean())
```

```python
md = pd.read_csv("Merged data.csv")
md
```

```python
md.insert(7, "wear", mean_wear)
md.insert(8, "friction", mean_friction)
md
```
