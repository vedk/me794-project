```python
import numpy as np
import pandas as pd
```

```python
n = 3 # number of replicates

s = 10
e = 60
```

```python
def bounds(r, s, e):
    return (s + r * (e - s + 1) // n, s + (r + 1) * (e - s + 1) // n - 1)
```

```python
mean_wear = []
mean_friction = []
for i in range(1, 97):
    df = pd.read_csv(f"Correct data/Piezo_{i}.dwf", sep=";", names=["timestamp", "wear", "friction", "const", "ignore"])
    
    wear_i = []
    friction_i = []
    for r in range(n):
        sr, er = bounds(r, s, e)
        wear_i.append(df["wear"].loc[sr:er].mean())
        friction_i.append(df["friction"].loc[sr:er].mean())

    mean_wear.append(wear_i)
    mean_friction.append(friction_i)
```

```python
weardf = pd.DataFrame(mean_wear, columns=[f"wear{r}" for r in range(n)])
weardf
```

```python
frictiondf = pd.DataFrame(mean_friction, columns=[f"friction{r}" for r in range(n)])
frictiondf
```

```python
md = pd.read_csv("Merged data.csv")
md
```

```python
df = md[["Sample Number", "Lubrication Condition", "Pin Material", "Weight (kg)", "RPM"]]
finaldf = pd.concat([df, weardf, frictiondf], axis=1)
```
