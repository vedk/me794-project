```r
df <- read.csv("Merged data draft 3 rows 10 to 60.csv")
df
```

```r
brass <- subset(df, Pin_Material == "Brass" & Weight == 9)
brass
```

```r
pin_materials <- c("Brass", "Aluminium")
y_axis_labels <- c("wear", "friction")
weights <- list(Brass = 9, Aluminium = 3)
x <- c(100, 150, 200, 250)
for (pin_material in pin_materials) {
  for (y_axis_label in y_axis_labels) {
    ss <- subset(
      df,
      Pin_Material == pin_material & Weight == weights[[pin_material]]
    )

    y <- matrix(
      c(
        subset(ss, Lubrication_Condition == "No Lubrication")[[y_axis_label]],
        subset(ss, Lubrication_Condition == "Half Lubrication")[[y_axis_label]],
        subset(ss, Lubrication_Condition == "Full Lubrication")[[y_axis_label]]
      ),
      nrow = 4,
      ncol = 3,
    )

    title <- paste(
      y_axis_label, "vs RPM for", pin_material,
      "weight =", weights[[pin_material]], "kg"
    )
    png(
      filename = paste0(title, ".png"),
      width = 1920, height = 1920, res = 288
    )
    matplot(
      x, y,
      type = "b", pch = 15:17, xlab = "RPM", ylab = y_axis_label,
      main = title, lty = 1, col = 1
    )
    legend("left", legend = c("No", "Half", "Full"), pch = 15:17)
    dev.off()
  }
}
```

```r
y_axis_labels <- c("wear", "friction")
x <- c("Brass", "Aluminium")
rpm <- 200
weights <- list(Brass = 9, Aluminium = 3)

for (y_axis_label in y_axis_labels) {
  y <- matrix(
    c(
      subset(
        df,
        Pin_Material == x[1] & RPM == rpm & Weight == weights[[x[1]]]
      )[[y_axis_label]],
      subset(
        df,
        Pin_Material == x[2] & RPM == rpm & Weight == weights[[x[2]]]
      )[[y_axis_label]]
    ),
    nrow = 3,
    ncol = 2
  )

  title <- paste(
    y_axis_label,
    "vs material for RPM =",
    rpm
  )
  png(
    filename = paste0(title, ".png"),
    width = 1920, height = 1920, res = 288
  )
  barplot(
    y,
    beside = TRUE, xlab = "Material", ylab = y_axis_label,
    names.arg = c("Brass 9 kg", "Aluminium 3 kg"), main = title,
    args.legend = list(x = "topright"), legend.text = c("No", "Half", "Full")
  )
  dev.off()
}
```

```r
y_axis_labels <- c("wear", "friction")
x <- c("No Lubrication", "Half Lubrication", "Full Lubrication")
rpm <- 200
weights <- list(Brass = 9, Aluminium = 3)

for (y_axis_label in y_axis_labels) {
  y <- matrix(
    c(
      subset(
        df,
        Lubrication_Condition == "No Lubrication" & RPM == 200 &
          (Weight == weights[["Brass"]] | Weight == weights[["Aluminium"]])
      )[[y_axis_label]],
      subset(
        df,
        Lubrication_Condition == "Half Lubrication" & RPM == 200 &
          (Weight == weights[["Brass"]] | Weight == weights[["Aluminium"]])
      )[[y_axis_label]],
      subset(
        df,
        Lubrication_Condition == "Full Lubrication" & RPM == 200 &
          (Weight == weights[["Brass"]] | Weight == weights[["Aluminium"]])
      )[[y_axis_label]]
    ),
    nrow = 2,
    ncol = 3
  )

  title <- paste(
    y_axis_label,
    "vs lubrication for RPM =",
    rpm
  )
  png(
    filename = paste0(title, ".png"),
    width = 1920, height = 1920, res = 288
  )
  barplot(
    y,
    beside = TRUE, xlab = "Lubrication Condition", ylab = y_axis_label,
    names.arg = x, main = title, args.legend = list(x = "topright"),
    legend.text = c("Brass 9 kg", "Aluminium 3 kg")
  )
  dev.off()
}
```
