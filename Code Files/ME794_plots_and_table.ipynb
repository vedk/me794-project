```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('Test data 10 to 60 3 replicates.csv')
```

```python
# Create four subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot friction vs Lubrication Condition
df.plot(x='Lubrication Condition', y='friction', ax=axes[0, 0])
axes[0, 0].set_title('Friction vs Lubrication Condition')

# Plot friction vs Pin Material
df.plot(x='Pin Material', y='friction', ax=axes[0, 1])
axes[0, 1].set_title('Friction vs Pin Material')

# Plot friction vs Weight (kg)
df.plot(x='Weight (kg)', y='friction', ax=axes[1, 0])
axes[1, 0].set_title('Friction vs Weight (kg)')

# Plot friction vs RPM
df.plot(x='RPM', y='friction', ax=axes[1, 1])
axes[1, 1].set_title('Friction vs RPM')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Create four subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot friction vs Lubrication Condition
df.plot(x='Lubrication Condition', y='wear', ax=axes[0, 0])
axes[0, 0].set_title('Wear vs Lubrication Condition')

# Plot friction vs Pin Material
df.plot(x='Pin Material', y='wear', ax=axes[0, 1])
axes[0, 1].set_title('Wear vs Pin Material')

# Plot friction vs Weight (kg)
df.plot(x='Weight (kg)', y='wear', ax=axes[1, 0])
axes[1, 0].set_title('Wear vs Weight (kg)')

# Plot friction vs RPM
df.plot(x='RPM', y='wear', ax=axes[1, 1])
axes[1, 1].set_title('Wear vs RPM')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Create a scatter plot of friction vs wear
plt.scatter(df['friction'], df['wear'])

# Set the title and axis labels
plt.title('Friction vs Wear')
plt.xlabel('Friction')
plt.ylabel('Wear')

# Show the plot
plt.show()
```

```python
# Group the data by Lubrication Condition, Pin Material, Weight (kg), and RPM
grouped_data = df.groupby(['Lubrication Condition', 'Pin Material', 'Weight (kg)', 'RPM'])

# Count the number of rows in each group
counts = grouped_data.size().unstack()

# Create a table of the counts
print(counts.to_string())
```
